# Type I and Type II Errors



Type I error is False Positive and type II error is False Negative
Type I error is when null hypothesis is true but is falsly rejected and Type II error is when null hypothesis is not true and we errorneously accept it.

Let's imagine that we are doing a test to check if a cell is cancerous or not. The null hypothesis is that "The cell is not cancerous"
Type I error happens when we have a cell that is healthy, so the null hypothesis is true but we reject it and classify the cell as cancerous. Therefore it is a false positive.

Type II error happens when we have a cancerous cell and we must reject the null hypothesis but we accept it. Therefore it is a false negative





## How are type I error and Type II error related?

To answere this question let's give an example.
Imagine that the distribution of diameter of healthy cells has a mean of 5 and standard deviation of 0.8. When the significance level (alpha) is 0.05, a healthy cell with daimeter bigger than the 95% percentile is classified as cancerous. However, it s a healthy cell. So it's a false positive.

In Type two error the alternative distribution is the true population distribution. However, we are working with the wrong distribution. Let's say that the population has a mean of 8 and standard deviation of 0.8. The distribution we working with has a mean of 5 and std of 0.8. 

Case I: alpha is 0.05

In this case anthing smaller than the critical line of 0.95 is healthy

```python
import numpy as np

# Import SciPy stats and matplotlib for calculating and visualising effect size
import scipy.stats
from scipy.stats import norm

import matplotlib.pyplot as plt
plt.style.use('seaborn')
import seaborn as sns
%matplotlib inline

# seed the random number generator so you get the same results
np.random.seed(42)
```

```python
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(20,6), sharey=True)
mean0 = 5
std0 = 0.8
x0 = np.linspace(norm.ppf(0.0001, mean0, std0), norm.ppf(0.9999, mean0, std0), 100)
y0 = norm.pdf(x0, mean0, std0)
ax0.plot(x0, y0, alpha=0.8, label='Null H', linewidth=4, color='green')
y_95 = norm.ppf(.95, mean0, std0)
ax0.vlines(y_95,ymin=0.01,ymax=0.5, linestyles ='dashed')
#ax0.vlines(norm.ppf(.05, mean0, std0), ymin=0.01, ymax=.35, linestyle='dashed')
ax0.fill_betweenx(y0,x1=x0,x2=y_95, where = x0 > y_95,color='r')

mean1 = 8
std1 = 0.8
x1 = np.linspace(norm.ppf(0.0001, mean1, std1), norm.ppf(0.9999, mean1, std1), 100)
y1 = norm.pdf(x1, mean1, std1)
ax0.plot(x1, y1, alpha=0.8, label='ALTERNATE H', linewidth=4, color='black')
ax0.fill_betweenx(y0, x1=x1, x2=y_95, where = x1 < y_95, color = 'b')
ax0.set_title('alpha = 0.05')
ax0.set_xlabel('Diameter (mm)')
ax0.legend(['Null Hypothesis', 'Alternative Hypothesis'], loc='upper right')


mean0 = 5
std0 = 0.8
x0 = np.linspace(norm.ppf(0.0001, mean0, std0), norm.ppf(0.9999, mean0, std0), 100)
y0 = norm.pdf(x0, mean0, std0)
ax1.plot(x0, y0, alpha=0.8, label='Null H', linewidth=4, color='green')
y_90 = norm.ppf(.90, mean0, std0)
ax1.vlines(y_90,ymin=0.01,ymax=0.5, linestyles ='dashed')
#ax0.vlines(norm.ppf(.05, mean0, std0), ymin=0.01, ymax=.35, linestyle='dashed')
ax1.fill_betweenx(y0,x1=x0,x2=y_90, where = x0 > y_90,color='r')


mean1 = 8
std1 = 0.8
x1 = np.linspace(norm.ppf(0.0001, mean1, std1), norm.ppf(0.9999, mean1, std1), 100)
y1 = norm.pdf(x1, mean1, std1)
ax1.plot(x1, y1, alpha=0.8, label='ALTERNATE H', linewidth=4, color='black')
ax1.fill_betweenx(y0, x1=x1, x2=y_90, where = x1 < y_90, color = 'b')
ax1.set_title('alpha = 0.1')
ax1.set_xlabel('Diameter (mm)')
ax1.legend(['Null Hypothesis', 'Alternative Hypothesis'], loc='upper right')


```
